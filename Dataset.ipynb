```python

orders = pd.read_csv('List_of_Orders_55FFC79CF8.csv')

#understanding the data
print(orders.head())
print(orders.describe())
print(orders.info())
print(orders.columns)
print(orders.shape)
print(orders.isnull().sum())
print(orders.dtypes)

```

# a. Orders by State
A bar chart showing the number of orders per state:

```python
plt.figure(figsize=(12, 6))
sns.countplot(y=orders["State"], order=orders["State"].value_counts().index, palette="coolwarm")
plt.title("Number of Orders per State")
plt.xlabel("Order Count")
plt.ylabel("State")
plt.show()
```

# b. Orders by City
A similar visualization but at the city level:

```python
plt.figure(figsize=(12, 6))
sns.countplot(y=orders["City"], order=orders["City"].value_counts().index[:15], palette="viridis")  # Top 15 cities
plt.title("Top 15 Cities with Most Orders")
plt.xlabel("Order Count")
plt.ylabel("City")
plt.show()
```

# c. Orders Over Time
If the Order Date column exists, we can analyze the order trend:

```python
orders["Order Date"] = pd.to_datetime(orders["Order Date"], format="%d-%m-%Y")  # Convert to datetime
orders_by_date = orders.groupby("Order Date").size()

plt.figure(figsize=(12, 6))
orders_by_date.plot(color="blue", marker="o", linestyle="-")
plt.title("Daily Order Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python

orders = pd.read_csv('Order_Details_19795F61CF.csv')

#understanding the data
print(orders.head())
print(orders.describe())
print(orders.info())
print(orders.columns)
print(orders.shape)
print(orders.isnull().sum())
print(orders.dtypes)
```

# a. Sales Distribution (Histogram)

```python
plt.figure(figsize=(12, 6))
sns.countplot(y=orders["State"], order=orders["State"].value_counts().index, palette="coolwarm")
plt.title("Number of Orders per State")
plt.xlabel("Order Count")
plt.ylabel("State")
plt.show()
```

# b. Profit Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(orders["Profit"], bins=30, kde=True, color="green")
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

# c. Sales by Category

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=orders["Category"].value_counts().index, y=orders["Category"].value_counts().values, palette="coolwarm")
plt.title("Number of Orders by Category")
plt.xlabel("Category")
plt.ylabel("Order Count")
plt.grid()
plt.show()
```

# e. Sales vs. Profit (Scatter Plot)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=orders["Amount"], y=orders["Profit"], hue=orders["Category"], palette="Set1", alpha=0.7)
plt.title("Sales Amount vs. Profit")
plt.xlabel("Sales Amount")
plt.ylabel("Profit")
plt.axhline(0, color="red", linestyle="--")  # Indicating break-even point
plt.grid()
plt.show()
```

# f. Sales Contribution per Category (Pie Chart)

```python
plt.figure(figsize=(8, 8))
orders.groupby("Category")["Amount"].sum().plot(kind="pie", autopct="%1.1f%%", cmap="coolwarm", startangle=140)
plt.title("Sales Contribution per Category")
plt.ylabel("")
plt.show()
```

# Sales Target

```python

# Load dataset
sales_target = pd.read_csv("Sales_target_DD2E9B96A0.csv")

# Convert 'Month of Order Date' to datetime format
sales_target["Month of Order Date"] = pd.to_datetime(sales_target["Month of Order Date"], format="%b-%y")

# Remove commas from 'Target' and convert to numeric
sales_target["Target"] = sales_target["Target"].replace(",", "", regex=True).astype(float)

# Display basic info
print(sales_target.head())
print(sales_target.describe())
print(sales_target.info())
print(sales_target.columns)
print(sales_target.shape)
print(sales_target.isnull().sum())
print(sales_target.dtypes)
```

# a. Sales Target Trend Over Time (Line Plot)

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=sales_target["Month of Order Date"], y=sales_target["Target"], marker="o", color="blue")
plt.title("Sales Target Over Time")
plt.xlabel("Month")
plt.ylabel("Target Sales Amount")
plt.grid()
plt.show()
```

# b. Monthly Target Distribution (Box Plot)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=sales_target, 
    x="Category", 
    y="Target", 
    hue="Category",  # Explicitly setting hue
    palette="Set2"
)
plt.title("Distribution of Monthly Sales Targets by Category")
plt.xlabel("Category")
plt.ylabel("Target Sales Amount")
plt.legend([], [], frameon=False)  # Hides redundant legend
plt.grid()
plt.show()
```

# c. Target Growth Rate Over Time (Percentage Change)

```python
sales_target["Target Growth (%)"] = sales_target["Target"].pct_change() * 100

plt.figure(figsize=(12, 6))
sns.lineplot(x=sales_target["Month of Order Date"], y=sales_target["Target Growth (%)"], marker="o", color="green")
plt.title("Target Growth Rate Over Time")
plt.xlabel("Month")
plt.ylabel("Growth Rate (%)")
plt.axhline(0, color="red", linestyle="--")  # Zero growth reference
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
